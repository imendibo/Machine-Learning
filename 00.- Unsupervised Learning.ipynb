```python
from sklearn.datasets.samples_generator import make_blobs, make_circles, make_moons
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, SpectralClustering
from sklearn.metrics import silhouette_score
%matplotlib widget
%matplotlib inline 
```

# Data

## Load

```python
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```


## Visualize

```python
plt.scatter(X[:, 0], X[:, 1], s=50, c=y)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
```

# K-means
http://alekseynp.com/viz/k-means.html
<img src="images/kmeans.gif" width="800">

```python
#number of clusters to create
k = 2

kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Category Number of Blobs")
```

## Selecting K value: Elbow method & Silhouette method

```python
# Run the Kmeans algorithm and get the index of data points clusters
sse = []
silhouette = []
# We will proof k values from 1 to 10
list_k_elb = list(range(1, 10))

for k in list_k_elb:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

list_k_sil = list(range(2, 10))

for k in list_k_sil:
    kmeans = KMeans(n_clusters=k).fit(X)
    score = silhouette_score(X, kmeans.predict(X), metric='euclidean')
    silhouette.append(score)

# Plot sse against k
plt.figure(figsize=(15, 6))
plt.plot(list_k_elb, sse, '-o')
plt.title("Elbow method")
plt.xlabel(r'Number of clusters K')
plt.ylabel('Sum of squared distance')

plt.figure(figsize=(15, 6))

plt.plot(list_k_sil, silhouette, '-o')
plt.title("Silhouette method")
plt.xlabel(r'Number of clusters K')
plt.ylabel('Average silhouette width')
```

**Elbow:**

As the sum of squared error is starting to converge in the 3rd cluster we will create only three clusters to sepparate the data. This K number selection technique is called the elbow method, because of the graph form and the interest point.

**Silhouette**:

As the average silhoutte with has its maximum in the 3rd cluster we will create three clusters.


## Clustering

```python
#number of clusters to create: selected from the output of the elbow and silhouette methods.
k = 3
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

## Plot result

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Correct Category Number of Blobs")
```

```python
#number of clusters to create
k = 3

kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(X_circles)
y_pred = kmeans.predict(X_circles)
```

## Kernel Transformation
We used a kernel transformation to **project the data into a higher dimension where a linear separation is possible**.


One version of this kernelized k-means is implemented in Scikit-Learn within the SpectralClustering estimator. It uses the graph of nearest neighbors to compute a higher-dimensional representation of the data, and then assigns labels using a k-means algorithm.

```python
sp = SpectralClustering(n_clusters=2, affinity='nearest_neighbors')
sp.fit(X_circles)
labels = sp.labels_

plt.scatter(X_circles[:, 0], X_circles[:, 1], c=labels)
```

# PCA: Principal Component Analysis
Problem: you have a multidimensional set of data and you want to see which points are closest to others.

PCA allows you to identify the dimensions of greatest variance, to the dimensions of least variance. PCA1 has greatest variance, PCA2 less and so on.

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d
```

## Data

```python
iris = datasets.load_iris()
iris.data.shape
```

As we can observe, iris dataset has 4 dimensions. In order to represent a plot of 2 dimensions we have to reduce the dimensionality.

## PCA 2D

```python
pca = PCA(n_components=2)

principalComponents = pca.fit_transform(iris.data)
```

```python
plt.scatter(principalComponents[:, 0], principalComponents[:, 1], c=iris.target)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
```

## PCA 3D

```python
pca = PCA(n_components=3)

principalComponents = pca.fit_transform(iris.data)
```

```python
fig = plt.figure()
plt.figure(figsize=(15, 10))

ax = plt.axes(projection='3d')
ax.scatter3D(principalComponents[:,0], principalComponents[:,1], principalComponents[:,2],c=iris.target)
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
```
